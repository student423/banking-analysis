```python
# create an income band from estimate income
bins = [0,100000,300000, float('inf')]
labels = ['Low Income', 'Medium Income', 'High Income']

df['Income band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels,right=False)
```

```python
# create and visualize in bar

df['Income band'].value_counts().plot(kind='bar')
```

```python
#examine the distribution of unique categories in categorical columns
categorical_col=df[['BRId',	'GenderId',	'IAId','Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure','Loyalty Classification', 'Properties Owned',
       'Risk Weighting','Income band']].columns

for col in categorical_col:
  print(f"value counts for '{col}':")
  display(df[col].value_counts())

```

### univariate analysis

```python
#visualize for better understand

for i, predictor in enumerate(df[categorical_col].columns):
    plt.figure(i, figsize=(8, 4))
    sns.countplot(data=df, x=predictor)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### bivariate analysis

```python
for i, predictor in enumerate(df[categorical_col].columns):
    plt.figure(i, figsize=(8, 4))
    sns.countplot(data=df, x=predictor,hue='GenderId',palette='Set2')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
for i, predictor in enumerate(df[categorical_col].columns):
    plt.figure(i, figsize=(8, 4))
    sns.countplot(data=df, x=predictor,hue='Nationality',palette='dark')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#histplot of value counts for different occupation

for col in categorical_col:
  if col=="Occupation":
     continue
  sns.set(rc={'figure.figsize':(8,4)})
  sns.histplot(df[col])
  plt.title("Histogram of occupation count")
  plt.tight_layout()
  plt.show()
```

### numerical analysis

```python
numerical_col=['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans','Foreign Currency Account', 'Business Lending',
       'Bank Deposits', 'Checking Accounts', 'Saving Accounts']

 # univariate analysis and visualization

sns.set(rc={'figure.figsize':(15,7)})
for i,col in enumerate(numerical_col):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
  plt.tight_layout()
  plt.show()
```

Heatmaps

```python
correlation_matrix=df[numerical_col].corr()
sns.set(rc={'figure.figsize':(12,10)})
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt=".2f",linewidths=0.5)
plt.tight_layout()
plt.show()
```

## Insights of EDA ⁉
🔹 Strong positive correlation (r = 0.84) observed between Bank Deposits and Checking Accounts, suggesting active depositors prefer easy access to funds.

🔸 Estimated Income shows weak correlation with most banking variables, indicating income alone doesn’t predict product usage.

🔹 Moderate correlation between Bank Loans and Business Lending (r = 0.42) implies overlap between personal and business credit users.
